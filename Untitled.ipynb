Моя модель с учетом того, что исходные данные были прологарифмированы, к итоговым предсказаниям нужно применить экспоненту: $$ y = X\theta + \epsilon, \epsilon \sim N(0, \sigma^2 I)$$
<br>Выберу семейство априорных распределений для параметров модели. В качестве семейства будет выступать Нормальое и обратное Гамма распределения (NIG) для наших параметров $(\theta, \sigma^2)$.
<br> <center>$\theta|\sigma^2 \sim N(\theta_{0}, \sigma^2V_{0})$ и $\sigma^2 \sim IG(\alpha_{0}, \beta_{0})$ </center>
$$(\theta, \sigma^2) \sim NIG(\theta_{0}, V_{0}, \alpha_{0}, \beta_{0})$$
<br>Прежде всего построю регрессию на первых 50 днях. Модель регрессии поможет мне определиться с исходными значениями параметров распределений. Для параметра $\sigma^2$ я сделаю оценку по выборке и предсказанию модели. Далее я выберу параметры $\alpha_{0}, \beta_{0}$ таким образом, чтобы получившееся распределение имело мат ожидание в значении оценки $\sigma^2$. 

# 2 Линейная регрессия, получение априорных распределений

```python
model = sm.OLS(data['log_total_cases'][:50], data[['day', 'ones']][:50]).fit()
model.summary()
```

```python
predictions_50 = model.predict()
predictions_over_50 = model.predict(data[['day', 'ones']][49:60])
```

```python
plt.figure(figsize=(14, 8))
plt.scatter(data['date'][:60], data['log_total_cases'][:60])
plt.plot(data['date'][:50], predictions_50, label='train data predictions', linewidth=2.5)
plt.plot(data['date'][49:60], predictions_over_50, label='out of train data')
plt.ylabel('log_total_cases')
plt.xticks( rotation=90)
plt.legend()
plt.grid()
plt.show()
```

```python
sigma = np.mean((predictions_50 - data['log_total_cases'][:50]) ** 2)
sigma
```

$\sigma^2 = \frac{\beta}{\alpha - 1}$ lets $\beta = 1$, then $\alpha = \frac{1}{\sigma^2} + 1$

```python
beta = 1
alpha = 1 / sigma + 1
alpha
```

$V_{0} = diag(v_{11}, v_{22})$ где $v_{ii} = 1$ 

```python
V = np.diag([1, 1])
```

Итого параметры априорого распределения таковы: $$\alpha_{0}  = 9.21, \beta_{0} = 1, (\theta_{0})_{1} = 0.1983, (\theta_{0})_{2} = 1.6224, (V_{0})_{11} = 1, (V_{0})_{22} = 1, (V_{0})_{12} = 0$$

Для итеративного обновления весов воспользуемся следующими формулами:
$$\theta_{1} = (V_{0}^{-1} + X^TX)^{-1}(V_{0}^{-1} + X^Ty) $$
$$V_1 = (V_{0}^{-1} + X^TX)^{-1} $$
$$\alpha_1 = \alpha_0 + \frac{1}{2}n $$
$$ \beta_1 = \beta_0 + \frac{1}{2}(\theta_0^TV_0^{-1}\theta_0 + y^Ty - \theta_1^TV_1^{-1}\theta_1)$$

# Апостериорное распределение 

```python
x_new = data[50:60][['day', 'ones']].values
y_new = data[50:60][['log_total_cases']].values
```

```python
theta = model.params.values.reshape((2, 1))
theta
```

```python
def bayesian_update(X, y, theta, V, alpha, beta):
    """
    X.shape = (n, m)
    y.shape = (n, 1)
    theta.shape = (m, 1)
    V.shape = (m, m)
    alpha is scalar
    beta is scalar
        n is number of obs
        m is number of features
    """
    inv_V = np.linalg.inv(V)
    theta_1 = np.linalg.inv((inv_V + X.T @ X)) @ (inv_V @ theta + X.T @ y)
    
    V_1 = np.linalg.inv((inv_V + X.T @ X))
    alpha_1 = alpha + 0.5 * X.shape[0]
    
    inv_V_1 = np.linalg.inv(V_1)

    beta_1 = beta + 0.5 * (theta.T @ inv_V @ theta + y.T @ y - theta_1.T @ inv_V_1 @ theta_1)
    return theta_1, V_1, alpha_1, beta_1[0][0]
```

```python
theta_1, V_1, alpha_1, beta_1 = bayesian_update(x_new, y_new, theta, V, alpha, beta)
sigma_1 = beta_1 / (alpha_1  - 1)
```

Итого параметры априорого распределения таковы: $$\alpha_{0}  = 9.21, \beta_{0} = 1, (\theta_{0})_{1} = 0.1983, (\theta_{0})_{2} = 1.6224, (V_{0})_{11} = 1, (V_{0})_{22} = 1, (V_{0})_{12} = 0$$
Итого параметры апостериорного распределения таковы: $$\alpha_{1}  = 14.2, \beta_{1} = 1.39, (\theta_{1})_{1} = 0.1719, (\theta_{1})_{2} = 1.7656, (V_{1})_{11} = 0.0003, (V_{1})_{22} = 0.9736, (V_{1})_{12} = -0.0175$$

```python
def normal_pdf(mu, sigma):
    x = np.linspace(stats.norm.ppf(0.0001, loc=mu, scale=sigma),
                    stats.norm.ppf(0.9999, loc=mu, scale=sigma), 100
    )
    normal = stats.norm.pdf(x, loc=mu, scale=sigma)
    return(x, normal)

def inverse_gamma_pdf(alpha, beta, q0, q1):
    x = np.linspace(stats.invgamma.ppf(q0, alpha, scale=1),
                    stats.invgamma.ppf(q1, alpha, scale=1), 100
    )
    inverse_gamma = stats.invgamma.pdf(x / beta, alpha, scale=1) / beta
    return(x, inverse_gamma)


x_prior_theta_1, normal_prior_theta_1 = normal_pdf(theta[0], sigma)
x_prior_theta_2, normal_prior_theta_2 = normal_pdf(theta[1], sigma)

x_posterior_theta_1, normal_posterior_theta_1 = normal_pdf(theta_1[0], sigma_1)
x_posterior_theta_2, normal_posterior_theta_2 = normal_pdf(theta_1[1], sigma_1)

x_prior_sigma, ig_prior_sigma = inverse_gamma_pdf(alpha, beta, q0=0.001, q1=0.99)
x_posterior_sigma, ig_posterior_sigma = inverse_gamma_pdf(alpha_1, beta_1, q0=0.001, q1=0.9999)


fig, ax = plt.subplots(1, 3, figsize=(21, 5))

ax[0].plot(x_prior_theta_1, normal_prior_theta_1, label=f'prior theta_1')
ax[0].plot(x_posterior_theta_1, normal_posterior_theta_1, label=f'posterior theta_1')

ax[1].plot(x_prior_theta_2, normal_prior_theta_2, label=f'prior theta_2')
ax[1].plot(x_posterior_theta_2, normal_posterior_theta_2, label=f'posterior theta_2')

ax[2].plot(x_prior_sigma, ig_prior_sigma, label=f'prior sigma')
ax[2].plot(x_posterior_sigma, ig_posterior_sigma, label=f'posterior sigma')

x_labels = ['theta_1', 'theta_2', 'sigma']
for i, axis in enumerate(ax):
    axis.grid(True)
    axis.legend()
    axis.set_xlabel(x_labels[i])
    axis.set_ylabel('density')
plt.show()
```

```python
def model_bayes(theta, x):
    return x @ theta
```

```python
posterior_predictions = model_bayes(theta_1, data[['day', 'ones']][:60].values)

plt.figure(figsize=(14, 8))
plt.scatter(data['date'][:60], data['log_total_cases'][:60])
plt.plot(data['date'][:50], predictions_50, label='train data predictions', linewidth=2.5)
plt.plot(data['date'][49:60], predictions_over_50, label='out of train data')
plt.plot(data['date'][:60], posterior_predictions, label='posterior ')


plt.ylabel('log_total_cases')
plt.xticks( rotation=90)
plt.legend()
plt.grid()
plt.show()
```

```python
data['mle_predictions_log'] = model.predict(data[['day', 'ones']])
data['bayes_predictions_log'] = model_bayes(theta_1, data[['day', 'ones']].values)
data['mle_predictions'] = np.exp(data['mle_predictions_log']).astype(int)
data['bayes_predictions'] = np.exp(data['bayes_predictions_log']).astype(int)
```

# 3 Обучение сигмоидальной функции

Моя идея заключается в том, чтобы обучать теперь модель не на общем числе зараженных, а на новых случаях. <br>1. Обучается линейная регрессия <br> 2. Извлекаются параметры модели для гауссиана <br> 3. Берется CDF получившегося гауссиана <br> 4. Масштабируется высота на общее число случаев заражений

```python
model_3 = sm.OLS(data['new_cases'][:50], data[['day']][:50]).fit()
model_3.summary()
```

```python
predictions_new_50 = model_3.predict()
predictions_new_over_50 = model_3.predict(data[['day']][49:60])

plt.figure(figsize=(14, 8))
plt.scatter(data['date'][:60], data['new_cases'][:60])
plt.plot(data['date'][:50], predictions_new_50, label='train data predictions', linewidth=2.5)
plt.plot(data['date'][49:60], predictions_new_over_50, label='out of train data')

plt.ylabel('new_cases')
plt.xticks( rotation=90)
plt.legend()
plt.grid()
plt.show()
```

```python
sigma = np.mean((predictions_new_50 - data['new_cases'][:50]) ** 2)
sigma
```

```python
def normal_cdf(mu, sigma, n):
    x = np.linspace(stats.norm.ppf(0.0001, loc=mu, scale=sigma),
                    stats.norm.ppf(0.9999, loc=mu, scale=sigma), n
    )
    normal = stats.norm.cdf(x, loc=mu, scale=sigma)
    return(x, normal)

```

```python
x_prior_new, normal_prior_new = normal_cdf(model_3.params[0], sigma, 50)

plt.figure(figsize=(14, 8))
plt.plot(data['date'][:50], data.iloc[50]['total_cases'] * normal_prior_new, label=f'model')
plt.plot(data['date'][:60], data['total_cases'][:60], label='total_cases')


plt.ylabel('total_cases')
plt.xticks( rotation=90)
plt.legend()
plt.grid()
plt.show()
```

```python
def inv(x):
    return 1 / x
def bayesian_update_1(X, y, theta, V, alpha, beta):
    """
    X.shape = (n, m)
    y.shape = (n, 1)
    theta.shape = (m, 1)
    V.shape = (m, m)
    alpha is scalar
    beta is scalar
        n is number of obs
        m is number of features
    """
    inv_V = inv(V)
    theta_1 = inv(inv_V + (X.T @ X)[0][0]) * (inv_V * theta + (X.T @ y)[0])
    
    V_1 = inv(inv_V + (X.T @ X)[0][0])
    alpha_1 = alpha + 0.5 * X.shape[0]
    
    inv_V_1 = inv(V_1)

    beta_1 = beta + 0.5 * (theta * inv_V * theta + y.T @ y - theta_1 * inv_V_1 * theta_1)
    return theta_1[0][0], V_1[0], alpha_1, beta_1[0][0]
```

```python
x_new = data['day'][50:60].values.reshape((-1, 1))
y_new = data['new_cases'][50:60].values
theta = model_3.params.values.reshape((1, -1))
V = np.asarray([1])
beta = 1
alpha = 1 / sigma + 1
alpha

theta_1, V_1, alpha_1, beta_1 = bayesian_update_1(x_new, y_new, theta, V, alpha, beta)
sigma_1 = beta_1 / (alpha_1  - 1)
```

```python
x_prior_new, normal_prior_new_posterior = normal_cdf(theta_1, sigma_1, 60)
plt.figure(figsize=(14, 8))
plt.plot(data['date'][:60], data.iloc[60]['total_cases'] * normal_prior_new_posterior, label=f'model posterior')
plt.plot(data['date'][:50], data.iloc[50]['total_cases'] * normal_prior_new, label=f'model prior')
plt.plot(data['date'][:60], data['total_cases'][:60], label='total_cases')


plt.ylabel('total_cases')
plt.xticks( rotation=90)
plt.legend()
plt.grid()
plt.show()
```
